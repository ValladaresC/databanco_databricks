#####Invocando librerias y funciones necesarias

```python
from pyspark.sql.functions import col
```

#####Cargando la tabla delta de la capa bronze

```python
trx_bronze_df = spark.table("catalgbanco.capabronze.trx_bronze")
```

##### Seleccionar todas las columnas del struct 'transaccion' y las columnas adicionales
##### De esta manera se genera un dataframe expandido

```python
df_expanded = trx_bronze_df.select(
    col("transaccion.autenticacion.metodo"),
    col("transaccion.autenticacion.exitoso"),
    col("transaccion.autenticacion.codigo_enviado"),
    col("transaccion.canal"),
    col("transaccion.cuenta_destino_app.banco_app"),
    col("transaccion.cuenta_destino_app.numero_cuenta_app"),
    col("transaccion.cuenta_destino_app.titular_app"),
    col("transaccion.cuenta_destino_web.banco_web"),
    col("transaccion.cuenta_destino_web.numero_cuenta_web"),
    col("transaccion.cuenta_destino_web.titular_web"),
    col("transaccion.descripcion"),
    col("transaccion.detalles_app.dispositivo_app"),
    col("transaccion.detalles_app.sistema_operativo"),
    col("transaccion.detalles_app.version_app"),
    col("transaccion.estado"),
    col("transaccion.fecha"),
    col("transaccion.id_transaccion"),
    col("transaccion.moneda"),
    col("transaccion.monto"),
    col("transaccion.tipo"),
    col("transaccion.ubicacion.latitud"),
    col("transaccion.ubicacion.longitud"),
    col("transaccion.detalles_web.ip_origen"),
    col("transaccion.detalles_web.navegador"),
    col("transaccion.detalles_web.dispositivo_web"),
    col("transaccion.ubicacion_atm.ciudad"),
    col("transaccion.ubicacion_atm.direccion"),
    col("transaccion.ubicacion_atm.codigo_atm"),
    col("transaccion.tarjeta.numero_tarjeta"),
    col("transaccion.tarjeta.titular_atm"),
    col("fecha_auditoria"),
    col("timestamp_auditoria")
)
```

#####Filtrando el dataframe expandido por las transacciones APP 
#####y borrando las columnas que no pertenecen a dicha transacción

```python
df_app = df_expanded.filter(col("canal") == "App").drop('codigo_enviado', 'banco_web', 'numero_cuenta_web', 'titular_web', 'ip_origen', 'navegador', 'dispositivo_web', 'ciudad', 'direccion', 'codigo_atm', 'numero_tarjeta','titular_atm')
```

#####Filtrando el dataframe expandido por las transacciones WEB 
#####y borrando las columnas que no pertenecen a dicha transacción

```python
df_web = df_expanded.filter(col("canal") == "Web").drop('exitoso', 'banco_app', 'numero_cuenta_app', 'titular_app', 'dispositivo_app', 'sistema_operativo', 'version_app', 'latitud', 'longitud', 'ciudad', 'direccion', 'codigo_atm', 'numero_tarjeta', 'titular_atm')
```

#####Filtrando el dataframe expandido por las transacciones ATM
#####y borrando las columnas que no pertenecen a dicha transacción

```python
df_atm = df_expanded.filter(col("canal") == "ATM").drop('metodo', 'exitoso', 'codigo_enviado', 'banco_app', 'numero_cuenta_app', 'titular_app', 'banco_web', 'numero_cuenta_web', 'titular_web', 'dispositivo_app', 'sistema_operativo', 'version_app', 'latitud', 'longitud', 'ip_origen', 'navegador', 'dispositivo_web')
```

#####Generando tabla delta para la transaccion APP en la capa Silver

```python
df_app.write.format("delta").mode("overwrite").saveAsTable("catalgbanco.capasilver.trx_app_silver")
```

#####Generando tabla delta para la transaccion WEB en la capa Silver

```python
df_web.write.format("delta").mode("overwrite").saveAsTable("catalgbanco.capasilver.trx_web_silver")
```

#####Generando tabla delta para la transaccion ATM en la capa Silver

```python
df_atm.write.format("delta").mode("overwrite").saveAsTable("catalgbanco.capasilver.trx_atm_silver")
```
