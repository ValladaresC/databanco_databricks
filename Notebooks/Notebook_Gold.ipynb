#### Resumen por canal, monto y fecha de transacci√≥n

```python
resumen = spark.sql ('''
                    SELECT canal, monto, DATE(fecha) AS fecha
                    FROM catalgbanco.capasilver.trx_app_silver
                    UNION ALL
                    SELECT canal, monto, DATE(fecha) AS fecha
                    FROM catalgbanco.capasilver.trx_web_silver
                    UNION ALL
                    SELECT canal, monto, DATE(fecha) AS fecha
                    FROM catalgbanco.capasilver.trx_atm_silver
                     ''')
display(resumen)
```

#### Agrupaci√≥n por canal y fecha de transacci√≥n

```python
resumen2 = spark.sql ('''
                    SELECT canal, round(SUM(monto),2) AS total_monto, DATE(fecha) AS fecha
                    FROM (
                        SELECT canal, monto, DATE(fecha) AS fecha
                        FROM catalgbanco.capasilver.trx_app_silver
                        UNION ALL
                        SELECT canal, monto, DATE(fecha) AS fecha
                        FROM catalgbanco.capasilver.trx_web_silver
                        UNION ALL
                        SELECT canal, monto, DATE(fecha) AS fecha
                        FROM catalgbanco.capasilver.trx_atm_silver
                    ) 
                    GROUP BY canal, fecha
                     ''')
display(resumen2)
```

#### Generaci√≥n de la tabla delta trx_resumen

```python
resumen.write.format("delta").mode("overwrite").saveAsTable("catalgbanco.capagold.trx_resumen")
```

#### Generaci√≥n de la tabla delta trx_resumen2

```python
resumen2.write.format("delta").mode("overwrite").saveAsTable("catalgbanco.capagold.trx_resumen2")
```
